### ðŸŒ» **Q4. What is the average salary by Department ?**

```python
df.head(2)
```

```python
# calculates the average salary for each department
df.groupby('Department')['Salary_INR'].mean()
```

```python
## group by department, calculate the mean salary, and convert the result from INR to thousands
dept = df.groupby('Department')['Salary_INR'].mean()/1000
dept
```

```python
## sort the calculated average salaries for each department in descending order
dept_sort = dept.sort_values(ascending=False)

dept_sort
```

```python
## create a bar plot to visualize the average salary by department
dept_sort.plot(x = dept.index, y = dept.values, kind = 'bar', color = 'salmon', legend = True , width = 0.8)
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

##ðŸŒ» **Q15. Identify departments with the highest attrition rate (Resigned %)**

```python
df.head(2)
```

```python
## groups the data by department to count total and resigned employees
dept_counts = df.groupby('Department')['Status'].agg(total_emp = 'count', resigned = lambda x: (x == 'Resigned').sum())

dept_counts
```

```python
type(dept_counts)
```

```python
## calculate attrition rate

dept_counts['attrition_rate_%'] = (dept_counts['resigned'] / dept_counts['total_emp']) * 100
```

```python
dept_counts
```

```python
## sort by attrition rate (highest first)

dept_counts.sort_values("attrition_rate_%", ascending = False)
```
