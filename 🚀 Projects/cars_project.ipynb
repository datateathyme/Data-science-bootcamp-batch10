# üéØ **1. Instruction ( For Data Cleaning )**

- ‚úÖ Find all Null Value in the dataset. If there is any null
value in any column, then fill it with the mean of that column.

```python
## check for the total count of missing values in each column.
cars.isnull().sum()
```

```python
## fill na
cars['Cylinders'].fillna(cars['Cylinders'].mean(), inplace = True)
```

```python
cars.isnull().sum()
```

```python
cars.head()
```

```python
## fills NaN values in the specified columns with their respective means.
columns_to_impute = ['EngineSize', 'Horsepower', 'MPG_City', 'MPG_Highway', 'Weight', 'Wheelbase', 'Length' ]

for col in columns_to_impute:
    cars[col].fillna(cars[col].mean(), inplace=True)
```

```python
cars.isnull().sum()
```

```python
cars.head(2)
```

```python
## fill na column object
object_columns_to_impute = ['Make', 'Model', 'Type', 'Origin', 'DriveTrain', 'MSRP', 'Invoice' ]

for col in object_columns_to_impute:
    # Use mode() to get the most frequent value for the current column.
    cars[col].fillna(cars[col].mode()[0], inplace=True)
```

```python
cars.isnull().sum()
```

```python
cars.head(10)
```

# üéØ **4.Instruction ( Removing unwanted records )**
