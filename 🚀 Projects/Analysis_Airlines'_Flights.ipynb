### ðŸ“© **Q1. What are the airlines in the dataset, accompanied by their frequencies?**

```python
df.head()
```

```python
## checking how many airlines are in the dataset
df['airline'].nunique()
```

```python
## showing the names of the airlines in the dataset
df['airline'].unique()
```

```python
## showing all the airlines with their frequencies
df['airline'].value_counts()
```

```python
## showing all the airlines with their number of flights in bar chart
df['airline'].value_counts(ascending=True).plot.barh(color = ['salmon', 'lightblue'])

plt.title("Airlines with frequencies")
plt.xlabel("Number of flights")
plt.ylabel("Airline")
plt.show()
```

### ðŸ“© **Q2. Show Bar Graphs representing the Departure Time & Arrival Time**

```python
df.head()
```

```python
## showing the Departure time for the flights
df['departure_time'].value_counts()
```

```python
## showing the arrival time for the flights
df['arrival_time'].value_counts()
```

```python
## showing the departure time & arrival time the flights with their counts

plt.figure(figsize = (16, 4))

plt.subplot(1, 2, 1)

plt.bar( df['departure_time'].value_counts().index, df['departure_time'].value_counts().values, color=['salmon','lightblue'])
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1, 2, 2)

plt.bar( df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values, color=['pink','lightgreen'])
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.show()
```

### ðŸ“© **Q4. Does price varies with airlines ?**

```python
df.head()
```

```python
## grouping the airlines and checking their mean price
df.groupby('airline')['price'].mean().sort_values()
```

```python
## drawing a categorical plot showing the mean ticket price for each airline
sns.catplot(x= 'airline', y= 'price', kind = 'bar', palette='rocket', data=df, hue = 'class')

plt.show()
```

### ðŸ“© **Q5. How is the price affected when tickets are bought in just 1 or 2 days before departure?**

```python
df.head()
```

```python
## check days_lef unique value
df['days_left'].unique()
```

```python
## checking the mean ticket price for different days_left
day_left_mean_price = df.groupby('days_left')['price'].mean()
day_left_mean_price
```

```python
day_left_mean_price.head(10)
```

```python
day_left_mean_price.tail(10)
```

```python
## plot chart
sns.relplot(y='price', x = 'days_left', kind = 'line', data=df)

plt.show()
```

### ðŸ“© **Q6. How does the ticket price vary between Economy and Business class?**

```python
df.head()
```

```python
## uqinue value
df['class'].unique()
```

```python
## filter out the record with Economy Class
economy_class = df[ df['class'] == 'Economy']
economy_class
```

```python
## check mean price for Economy Class tickets
economy_class['price'].mean()
```

```python
## filter out the record with Business Class
business_class = df[ df['class'] == 'Business']
business_class
```

```python
## check mean price for Business Class tickets
business_class['price'].mean()
```

```python
result = business_class['price'].mean() - economy_class['price'].mean()
result
```
