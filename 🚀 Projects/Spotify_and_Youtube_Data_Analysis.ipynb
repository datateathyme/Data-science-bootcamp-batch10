## üì© **Import Module and Load the Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## read csv file
df = pd.read_csv('spotify_youtube_dataset.csv')
```

```python
df.tail()
```

```python
## to show the names of columns of the dataframe
df.columns
```

```python
## to get some basic information about the dataset
df.info()
```

```python
## removing the columns from the dataframe
df.drop(columns = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube'], inplace = True)
```

```python
df.info()
```

```python
## checking missing values counts in each column of the dataframe
df.isnull().sum()
```

```python
## filling the missing values with 0 in Likes & Commnets column
df[['Likes', 'Comments']] = df[['Likes', 'Comments']].fillna(0)
```

```python
## to check the count of missing values in each column
df.isnull().sum()
```

```python
## It drops the rows that contains all or any missing values
df.dropna(inplace = True)
```

```python
## to check the count of missing values in each column
df.isnull().sum()
```

## üåª **Q1) Top 10 Artists - with the Highest Views on YouTube?**

```python
top10_streams_tracks = df[['Track', 'Stream']].sort_values(by = 'Stream', ascending = False).head(10)

top10_streams_tracks
```

## üåª **Q3) What are the most common Album Types on Spotify? How many tracks belong to each album type?**

```python
df.head(2)
```

```python
## to check the unique values in a column
df['Album_type'].unique()
```

```python
## It shows all unique values with their counts in the column
al_type = df['Album_type'].value_counts()

al_type
```

```python
## create a Pie chart
plt.pie( al_type, labels = al_type.index, autopct='%1.1f%%', startangle=60, shadow=True, explode = (0.03,0.03,0.03))
plt.title('Most Common Album Types on Spotify')
plt.show()
```

## üåª **Q4) How do the Average Views, Likes, and Comments are compared between different Album Types?**

```python
df.head(2)
```

```python
## group the Album Type column, and show the mean of three columns
df2 = df.groupby('Album_type')[['Views', 'Likes', 'Comments']].mean().reset_index()
df2
```

```python
## melt - unpivot a dataframe
df_melted  = pd.melt( df2, id_vars = 'Album_type', var_name = "Attribute", value_name = 'Total' )

df_melted
```

```python
## create Bar Plot
plt.figure(figsize = (10, 5))
sns.barplot(data = df_melted, x = 'Album_type', y = 'Total', hue = 'Attribute');
```
