```python
## merge dataframe -- (join data)
left = {
    'key': [1, 2, 3, 4],
    'name': ['ploy', 'joe', 'jane', 'anna'],
    'age': [20, 30, 40, 50]
}

right = {
    'key': [1, 2, 3, 4],
    'city': ['bangkok', 'london', 'paris', 'tokyo'],
    'zip': [1001, 2501, 2094, 9802]
}

df_left = pd.DataFrame(left)
df_right = pd.DataFrame(right)
```
