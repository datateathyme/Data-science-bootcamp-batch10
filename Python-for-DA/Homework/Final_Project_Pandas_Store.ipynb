### **#TODO 06** - which Segment has the highest profit in 2018

### **#TODO 07** - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

```python
## filter date
filter_date = df[ (df['Order Date'] >= '2019-04-15') & (df['Order Date'] <= '2019-12-31') ]

filter_date = filter_date.sort_values(by='Order Date', ascending=True)

filter_date
```

```python
filter_date.groupby('State')['Sales'].sum().sort_values().head(5)
```

### **#TODO 08** - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%

```python
df_2019 = df[ df['Order Year'] == 2019 ]
df_west_central_2019 = df_2019[ (df_2019['Region'] == 'West') | (df['Region'] == 'Central') ]

df_west_central_2019.head(2)
```

```python
## proportion of total sales (%)
result_prop = df_west_central_2019['Sales'].sum() / df_2019['Sales'].sum() * 100

result_prop = round(result_prop, 2)

print(result_prop, '%')
```

### **#TODO 09** - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

```python
df_2019_2020 = df[ (df['Order Year'] >= 2019) & (df['Order Year'] <= 2020) ]

df_2019_2020.head()

```

```python
# Top 10 popular products in terms of number of orders during 2019-2020
top_10_orders_2019_2020 = df_2019_2020.groupby('Product Name')['Order ID'].count().sort_values(ascending=False).head(10)

print("Top 10 Popular Products (Number of Orders) in 2019-2020:")
display(top_10_orders_2019_2020)
```

```python
# Top 10 popular products total sales
top_10_sales_2019_2020 = df_2019_2020.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_10_sales_2019_2020
```
